```python
from bs4 import BeautifulSoup
from urllib.request import urlopen
import requests
```

```python
page = urlopen("https://feeds.feedburner.com/com/Yeor")
```

```python
parse = page.read()
```

```python
page.close()
```

```python
soup = BeautifulSoup(parse,"xml")
```

```python
items_list = soup.find_all("item")
```

```python
first = items_list[0]
```

```python
first
```

```python
link = first.find('feedburner:origLink')
```

```python
link.text
```

```python
news_page_to_follow = link.text
news_page_to_follow
```

```python
page_news = requests.get(news_page_to_follow).text
```

```python
news_soup = BeautifulSoup(page_news,'html.parser')
```

```python
#print(news_soup.prettify())
title = news_soup.find('h1')
title.text
```

```python
details = news_soup.find('div',class_='story-detail')
```

```python
all_the_para = details.find_all('p')
```

```python
all_the_para
```

```python
all_text = ""
for i in all_the_para:
    all_text += i.text
```
